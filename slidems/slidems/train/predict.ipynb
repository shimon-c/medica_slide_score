```python
from fastcore.all import *
from fastai.learner import load_learner
from fastai.vision.all import *
```

```python
#learn = load_learner("/home/dudi/privet/med/deeplearning/model/resize_192_squish_resnet18_epoc5.pkl")
learn = load_learner("/home/dudi/privet/med/deeplearning/model/resize_256_resnet50_epoc5.pkl")
```

```python
tssPath = Path("/home/dudi/privet/med/deeplearning/imgdb/train_set/GoodFocus/ANONFACHSI1RE_8_1_5_38.jpeg")
img = PILImage.create(tssPath)
className,classNum,probs = learn.predict(img)
print(f"This is a: {className}. with probability {probs[classNum]:.4f}")
img.to_thumb(500,500)
```
